### Black Friday Dataset EDA and Feature Engineering
#### Cleaning and Preparing data for Model Training

```python
### Dataset Link:  https://www.kaggle.com/datasets/sdolezel/black-friday
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Problem Statement
A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
# Import train data set
train_set = pd.read_csv('train.csv')
train_set.head()
```

```python
# Import test data set
test_set = pd.read_csv('test.csv')
test_set.head()
```

```python
# Merge both Train and Test Data
df = pd.concat([train_set,test_set])
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
pd.get_dummies(df['Gender'])
```

```python
# Handling Categorical Feature Gender
 
df['Gender'] = df['Gender'].map({'F': 0,'M':1})
df.head()
```

```python
# Handling Categorical Feature Age

df['Age'].unique()
```

```python
df['Age'] = df['Age'].map({'0-17': 1,'18-25':2 , '26-35' : 3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
##second technqiue
from sklearn import preprocessing
 
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
 
# Encode labels in column 'species'.
df['Age']= label_encoder.fit_transform(df['Age'])
 
df['Age'].unique()
```

```python
df.head()
```

```python
# Fixing Categorical Feature City_Category
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
df_city.head()
```

```python
df =pd.concat([df,df_city],axis=1)
df.head()
```

```python
df.drop('City_Category',axis=1,inplace=True)
df.head()
```

```python
# Missing Values
df.isnull().sum()
```

```python
# Focus on Replacing Product_Category_2 and Product_Category_3
# Purchase null value is because of Test Data

df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].value_counts()
```

```python
df['Product_Category_2'].mode()[0]
```

```python
# Replace NaN value of Product_Category_2 with Mode
df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
# Replace NaN value of Product_Category_3 with Mode
df['Product_Category_3'] = df['Product_Category_2'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df['Product_Category_3'].isnull().sum()
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.head()
```

```python
df.info()
```

```python
# Stay_In_Current_City_Years convert object into Integers
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(int)
```

```python

# Stay_In_Current_City_Years data type changed
df.info()
```

```python
# Change B and C data Type
df['B']=df['B'].astype(int)
df['C']=df['C'].astype(int)
```
