## Feature Engineering

```python
final_df["Date_of_Journey"].str.split("/").str[0] # Will get all Date in df
```

Derived Feature from Date_of_Journey

```python
# Process to get date and time
final_df["Date"] = final_df["Date_of_Journey"].str.split("/").str[0]
final_df["Month"] = final_df["Date_of_Journey"].str.split("/").str[1]
final_df["Year"] = final_df["Date_of_Journey"].str.split("/").str[2]
```

```python
final_df.head()
```

```python
# Convert Date into int
final_df["Date"] = final_df["Date"].astype(int)
final_df["Month"] = final_df["Month"].astype(int)
final_df["Year"] = final_df["Year"].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
final_df.head()
```

Now taking Arrival_Time remove Date and take only time

```python
# The same operation we applied above but this time using lambda for Arrival_Time
final_df["Arrival_Time"] = final_df["Arrival_Time"].apply(lambda x: x.split(" ")[0])
```

```python
final_df["Arrival_Hour"] = final_df["Arrival_Time"].apply(lambda x: x.split(":")[0])
final_df["Arrival_Min"] = final_df["Arrival_Time"].apply(lambda x: x.split(":")[1])
```

```python
# Convert Arrival_Time into int
final_df["Arrival_Hour"] = final_df["Arrival_Hour"].astype(int)
final_df["Arrival_Min"] = final_df["Arrival_Min"].astype(int)
```

```python
final_df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df.head(1)
```

Now Taking Departure Time

```python
final_df["Dep_Hour"] = final_df["Dep_Time"].apply(lambda x: x.split(":")[0])
final_df["Dep_Min"] = final_df["Dep_Time"].apply(lambda x: x.split(":")[1])
final_df["Dep_Hour"] = final_df["Dep_Hour"].astype(int)
final_df["Dep_Min"] =final_df["Dep_Min"].astype(int)
final_df.drop("Dep_Time",axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df.head(5)
```

Uptill now we removed Date_of_Journey, Arrival_Time and Dep_Time Features

Now Taking Total_Stops Feature

```python
final_df["Total_Stops"].unique()
```

```python
final_df[final_df["Total_Stops"].isnull()]
```

```python
# Change '1 stop' => 1 in df["Total_Stops"]
final_df["Total_Stops"] = final_df["Total_Stops"].map({"non-stop" : 0 ,"1 stop" : 1, "2 stops" : 2, "3 stops" : 3, "4 stops" : 4,np.nan : 1})
```

```python
final_df.head(5)
```

```python
# Now we drop Route Feature
final_df.drop("Route",axis=1,inplace=True)
```

```python
final_df.head(1)
```

Now Taking Duration Feature

```python
# In some Record Duration is 5m
final_df[final_df["Duration"] == "5m"]
```

```python
# Drop Rows with only duration in mins
final_df.drop(final_df[final_df["Duration"] == '5m'].index,inplace=True)
```

```python
# Convert Duration Hour into min
final_df['Duration_Min'] = final_df["Duration"].apply(lambda x: x.split(" ")[0].split("h")[0]).astype(int)

# Convert Duration hour into hour
final_df['Duration_Min'] = final_df['Duration_Min'] * 60
final_df["Duration_Min"]
```

```python
# Transforming Rows without min index into 0 Min
temp =  final_df["Duration"].apply(lambda x: x.split(" ")[1] if len(x.split(" ")) > 1 else 0)

# Removing 'm' from mins index
temp = temp.apply(lambda x:  x if x == 0 else x.split("m")[0])

# Change temp to int
temp = temp.astype(int)
temp
```

```python
# Add hour's min with min
final_df['Duration_Min'] = final_df['Duration_Min'] + temp
```

```python
final_df.columns.values
```

```python
final_df.head(2)
```

```python
# Remove 'Duration' column
final_df.drop("Duration",axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df.info()
```

```python
final_df["Additional_Info"].unique()
```

### Now, doing categorical feature

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
```

```python
final_df["Airline"] = labelencoder.fit_transform(final_df["Airline"])
final_df["Source"] = labelencoder.fit_transform(final_df["Source"])
final_df["Destination"] = labelencoder.fit_transform(final_df["Destination"])
final_df["Additional_Info"] = labelencoder.fit_transform(final_df["Additional_Info"])
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
final_df = pd.get_dummies(final_df,columns=["Airline", "Source", "Destination","Additional_Info"] ,drop_first = True)
```
