```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
# df_country.columns
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
# plt.legend()
plt.show()
```

Observation: Zomato most sales are from India followed by USA and United Kingdom

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
# sns.show()
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue='Rating color',palette=["blue","red","orange","yellow","green","green"])
```

Observation: 
1. Not Rated amount is very high
2. Maximum no.of ratings are between 2.4 to 3.4
